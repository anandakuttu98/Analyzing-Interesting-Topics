```python
import_table = pd.read_excel("Road Accidents.xlsx", sheet_name = "Sheet3", index_col = 0)
import_table.to_csv("Road Accidents.csv", encoding = 'utf-8', index = False)
import_table
```

```python
import_table = import_table.astype(int)
```

```python
import_table.info()
```

```python
import_table
```

```python
import_table.rename_axis(index = "States", columns = "Years", inplace = True)
import_table
```

```python
road_acd_rpt = import_table.iloc[:, :12] 
road_acd_rpt
```

```python
road_acd_ncrb = import_table.iloc[:, 12:]
road_acd_ncrb
```

```python
total_report = road_acd_rpt[-1:]
total_report
```

```python
total_ncrb = road_acd_ncrb[-1:]
total_ncrb
```

```python
total_report.T
```

```python
total_report.values
```

```python
total_ncrb.values
```

```python
new_data = pd.DataFrame(np.concatenate([total_report.values, total_ncrb.values], axis=0), 
                        index= ['Road Accident Report by Ministry','Road Accident Report by NCRB'], 
                        columns = ['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022'])
```

```python
new_data
```

```python
fig, ax = plt.subplots(figsize = (10, 10))
ax.plot(new_data.T.index,
       new_data.T)
ax.set(title = "Plotting the Trend of Road Accidents presented by the Ministry and NCRB over the years",
      xlabel = "Years",
      ylabel = "Total Cases of Road Accidents (in lakhs)")
ax.legend(new_data.T, loc = "upper right", title = "Trend Line")
fig.suptitle("Trend Line of the Number of Accidents per Year from 2011 to 2022",
            fontsize = 16,
            fontweight = "bold");
```
